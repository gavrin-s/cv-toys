```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image

from src.config import DATA_PATH
from src.utils import plot_images
```

```python
image = cv2.imread(path.join(DATA_PATH, "page.jpg"), 0)
Image.fromarray(image)
```

```python
ddepth = -1 # cv2.CV_64F
ksize = 3
figsize = (10, 10)

image_transformed_laplasian = cv2.Laplacian(image, ddepth)
image_transformed_sobelx = cv2.Sobel(image, ddepth, 1, 0, ksize=ksize)
image_transformed_sobely = cv2.Sobel(image, ddepth, 0, 1, ksize=ksize)
image_transformed_scharrx = cv2.Scharr(image, ddepth, 1, 0)
image_transformed_scharry = cv2.Scharr(image, ddepth, 0, 1)


plot_images([image, image_transformed_laplasian],
            names=["Original", "Laplasian"],
            figsize=figsize, imshow_kwargs={"cmap":plt.cm.gray}
           )
plt.show()

plot_images([image_transformed_sobelx, image_transformed_sobely],
            names=["Sobel X", "Sobel y"],
            figsize=figsize, imshow_kwargs={"cmap":plt.cm.gray}
           )
plt.show()

plot_images([image_transformed_scharrx, image_transformed_scharry],
            names=["Scharr X", "Scharr y"],
            figsize=figsize, imshow_kwargs={"cmap":plt.cm.gray}
           )
plt.show()
```

# Difference betwen uint and float for gradient calculation

```python
image = cv2.imread(path.join(DATA_PATH, "black_box.jpg"), 0)
Image.fromarray(image)
```

```python
ksize = 3
# uint8
image_transformed_sobelx = cv2.Sobel(image, ddepth, 1, 0, ksize=ksize)

# float64
image_transformed_sobelx_64f = cv2.Sobel(image, cv2.CV_64F, 1, 0, ksize=ksize)
image_transformed_sobelx_64f = np.absolute(image_transformed_sobelx_64f).astype(np.uint8)
# image_transformed_sobelx_64f = np.uint8(image_transformed_sobelx_64f)

plot_images([image_transformed_sobelx, image_transformed_sobelx_64f],
            names=["Sobel x uint8", "Sobel x float 32"],
            figsize=(20, 15), imshow_kwargs={"cmap":plt.cm.gray}
           )
```
