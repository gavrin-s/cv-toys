```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image

from src.config import DATA_PATH
```

```python
image = cv2.imread(path.join(DATA_PATH, "shrek.jpg"))
Image.fromarray(image)
```

```python
# Average filter
kernel = np.ones((3, 3), np.float32)/9
image_transformed = cv2.filter2D(image, -1, kernel)
Image.fromarray(image_transformed)
```

```python
# Gaussian filter
transformed_image = cv2.GaussianBlur(image, (3, 3), 0)
Image.fromarray(transformed_image)
```

```python
# Median filter
transformed_image = cv2.medianBlur(image, 5)
Image.fromarray(transformed_image)
```

```python
# bilateral filter
transformed_image = cv2.bilateralFilter(image, 5, 100, 100)
Image.fromarray(transformed_image)
```
